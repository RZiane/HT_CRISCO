```python
from lxml import etree
```

```python
path_xml = '/home/ziane212/crisco_work/ressources/guernesey/Guernesey_Crime_I_Sorcherie_SRCMF_convTags_7-2-23.Final.xml'                                         
#path_xml = '/home/ziane212/crisco_work/ressources/1862_histoire/1862_Histoire_de_la_Normandie_ancienne_H-T.xml'
path_dict = '/home/ziane212/crisco_work_ressources/test_dict.txt'
```

```python
list_upos = [("ADJ", "ADJ"),
             ("ADP", "PREP"),
             ("PUNCT", "PUNC"),
             ("ADV", "ADV"),
             ("SYM"),
             ("INTJ", "INT"),
             ("CCONJ", "CON"),
             ("X", "RES"),
             ("NOUN", "NOM"),
             ("DET", "DET"),
             ("PROPN", "NOMP"),
             ("NUM", "NUM"),
             ("PART"),
             ("PRON", "PRO"),
             ("SCONJ", "CON")]

def extract_dict(path_xml):
    l_dict = []
    
    d_dict = dict()
    
    tree = etree.parse(open(path_xml, encoding='utf-8'))

    for token in tree.findall('.//w'):

        if len(list(token))==0:
            form = token.text.replace('\t', '')
        else:
            if list(token)[0].tag == 'choice':
                form = list(token)[0][1].text.replace('\t', '')

        lemma = token.get('lemma')
        UPOS = token.get('udpos')
        UPENN_POS = token.get('uppos')
        PRESTO_FEAT = token.get('prpos')
        function = token.get('function')
        
        form = form.lstrip(' —').rstrip(' ')
        if UPOS!="PROPN":
            form = form.lower()
        
        # conversion POS PRESTO 
        if UPOS == 'AUX' or UPOS == 'VERB':
            if PRESTO_FEAT == 'Ge' or PRESTO_FEAT == 'Ga':
                PRESTO_POS = 'PAG'
            else:
                PRESTO_POS = 'VER'
        else:
            for POS in list_upos:
                if UPOS == POS[0]:
                    PRESTO_POS = POS[1]
        
        entry = form+'/'+PRESTO_POS+'/'+PRESTO_FEAT+'/'+lemma+'/'+UPOS+'/'+UPENN_POS
        functions = []
        
        if entry not in d_dict.keys():
            functions.append(function)
            d_dict[entry] = functions
        else:
            if function in d_dict[entry]:
                pass
            else:
                d_dict[entry].append(function)
                
    for k, v in d_dict.items():
        x = (k, ';'.join(list(sorted(set(v)))))
        l_dict.append(x)
                    
    l_dict = list(sorted(set(l_dict)))
    
    #print(d_dict)
    
    return l_dict

l_extraction = extract_dict(path_xml)
```

```python
with open(path_dict, 'r', encoding="utf-8") as dict_file:
    
    l_dict = []
    for i in dict_file:
        
        i = i.rstrip('\n')
        i = i.split('+')
        l_dict.append(i)
        
print(l_dict)

l_out = []
with open(path_dict, 'a', encoding="utf-8") as dict_file:
    
    if l_dict == []:
        for i in l_extraction:
            if i not in l_dict:
                dict_file.write('+'.join(i)+'\n')
    else:
        for i in l_extraction:
            if i[0] not in [ele for ele in l_dict][0]:
                l_out.append(i)
            elif i[0] == [ele for ele in l_dict][0][0] and i[1] != [ele for ele in l_dict][0][1]:
                i = (i[0], ';'.join(list(sorted(set(
                    [*i[1].split(';'), *[ele for ele in l_dict][0][1].split(';')])))))
                l_out.append(i)

        for i in l_out:
            dict_file.write('+'.join(i)+'\n')    
    
with open(path_dict, 'r') as r:
    l_orderedDict = list(sorted(set(r.readlines())))
    
with open(path_dict, 'w') as out:
    for i in l_orderedDict:
        out.write(i)

    
    
```
